## AgentQLLoader 

### AgentQLLoader requires the following two parameters:
### url: The URL of the web page you want to extract data from.
### query: The AgentQL query to execute. Learn more about how to write an AgentQL query in the docs or test one out in the AgentQL Playground. 




```python
import os

# Retrieve the AgentQL API key from environment variables
agentql_api_key = os.getenv("AGENTQL_API_KEY")
if agentql_api_key is not None:
	os.environ["AGENTQL_API_KEY"] = agentql_api_key  # Set only if found
else:
	print("Warning: AGENTQL_AsPI_KEY not found in environment variables. Please check your .env file if you need AgentQL features.")
```

```python
from langchain_agentql.document_loaders import AgentQLLoader
```

```python
agent_loader = AgentQLLoader(
    url = "https://www.linkedin.com/in/vaishnavi-kulange-556601278/",
    query= """ {
      posts[]{
      title
      url
      date
      author
      }
    }
    """, 
    is_scroll_to_bottom_enabled=True,

)
```

```python
agent_docs = agent_loader.load()
agent_docs[0]
```

```python
print(agent_docs[0].metadata)
```

## Amazon Textract

#### Amazon Textract 
#####   is a machine learning (ML) service that automatically extracts text, handwriting, and data from scanned documents. It goes beyond simple optical character recognition (OCR) to identify, understand, and extract data from forms and tables. Today, many companies manually extract data from scanned documents such as PDFs, images, tables, and forms, or through simple OCR software that requires manual configuration (which often must be updated when the form changes). To overcome these manual and expensive processes, Textract uses ML to read and process any type of document, accurately extracting text, handwriting, tables, and other data with no manual effort.

## UnstructuredFileLoader (Most Flexible)

```python
from langchain_community.document_loaders import UnstructuredFileLoader , DirectoryLoader
```

```python
from langchain_community.document_loaders import (
    DirectoryLoader,
    UnstructuredFileLoader,
)

loader = DirectoryLoader(
    "/Users/vaishnavirohitkulange/Documents/GENAI_Calssses/RAG_DEMO/data",
    glob="**/*.*",
    loader_cls=UnstructuredFileLoader,
)

docs = loader.load()

print(f"Loaded {len(docs)} documents")
```
