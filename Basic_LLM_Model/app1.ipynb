```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

load_dotenv()
```

```python

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant that answers questions based on the provided context. If you don't know the answer, say you don't know."),
        ("user", "Context: {context}\nQuestion: {question}")
    ]
)
```

```python
model = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
```

```python
output_parser = StrOutputParser()
```

```python
context = """ Artificial Intelligence (AI) has grown rapidly thanks to Deep Learning, 
a powerful approach where multi-layer neural networks learn patterns from massive amounts of data. 
This progress has transformed Natural Language Processing (NLP), allowing models like GPT, BERT, and LLaMA to 
understand and generate human language with impressive accuracy—handling tasks such as translation, summarization, 
and chat-based conversations. These breakthroughs were made possible by advances in computing power, large-scale datasets, 
and transformer architectures. Fun fact: the famous transformer model was introduced in a 2017 research paper titled “
Attention Is All You Need,” which has since become one of the most influential AI papers ever written and changed the 
entire field of NLP!"""
```

```python
question = "How many fruits you have ?"
```

```python
chain = prompt | model | output_parser
```

```python
output = chain.invoke({"context": scope, "question": question})
```
