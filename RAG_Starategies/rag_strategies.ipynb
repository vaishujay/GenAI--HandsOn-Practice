```python
## -----Configuration------- ##
load_dotenv()
```

```python
embedding_model = "text-embedding-ada-002"

llm_model = "gpt-3.5-turbo"

data_path = r"/Users/vaishnavirohitkulange/Documents/GENAI_Calssses/RAG_Starategies/Rainbow-Bazaar-Return-Refund-&-Cancellation-Policy.pdf"

chromadb_path = r"chroma_db"
```

```python
llm = ChatOpenAI(model_name = llm_model, temperature= 0)
embeddings = OpenAIEmbeddings(model = embedding_model)
```

## Load the document

```python
loader = PyPDFLoader(data_path)
pdf_doc = loader.load()
print(f"Number f pages in the document : {len(pdf_doc)}")
```

## Fixed Size Chunking (Fixed-Size (Character))

```python
# Join all page texts into a single string
all_text = "".join([doc.page_content for doc in pdf_doc])

# Character-based splitting
char_splitter = CharacterTextSplitter(chunk_size = 50 , 
                                     chunk_overlap = 10)

fixed_char_chunk = char_splitter.split_text(all_text)
print(fixed_char_chunk)
print(f"Number of chunks created: {len(fixed_char_chunk)}")

```

## Add vectors in vector DB 

```python
from langchain_core.documents import Document

# Convert string chunks to Document objects
fixed_char_docs = [Document(page_content=chunk) for chunk in fixed_char_chunk]

fixed_vectordb = Chroma.from_documents(fixed_char_docs, embeddings, persist_directory=chromadb_path)
```

## Text structure-based (Fixed Size chunking)


```python
fixed_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000 ,
                                                 chunk_overlap = 200)

fixed_chunk = fixed_splitter.split_documents(pdf_doc)
print(f"Number of fixed -size  chunks :{len(fixed_chunk)}")
```

## Recursive Chunking 

```python
recursive_splitter = RecursiveCharacterTextSplitter(chunk_size = 800 ,
                                                    chunk_overlap = 150,
                                                    separators= ['\n\n','\n'," "])

recursive_chunk = recursive_splitter.split_documents(pdf_doc)
print(f"Number of fixed size chunk : {len(recursive_chunk)}")
```
